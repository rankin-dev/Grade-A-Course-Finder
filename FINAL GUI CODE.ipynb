```python
import tkinter as tk
import tkinter.font as font
import csv

root = tk.Tk()
myFont = font.Font(family='Helvetica', size = 20, weight = 'bold')

def see_review():
    window = tk.Toplevel(root)
    with open("reviews.csv", newline = "") as file:
        reader = csv.reader(file)
        r = 0
        for col in reader:
            c = 0
            for row in col:
                label = tk.Label(window, width = 25, height = 2, \
                                   text = row, relief = tk.RIDGE)
                label.grid(row = r, column = c)
                c += 1
            r += 1
            
def create_review():
    window = tk.Toplevel(root)

    def add_to_csv():
        def user_input():
        #creates the pandas dataframe and uses user input for the values
            user_input_list = [e1.get(), e2.get(), e3.get(), e4.get(), e5.get(), e6.get(),e7.get()]
            return user_input_list

        def next_review():
            with open("reviews.csv", "a") as new_file:
                #creates a writing variable
                csv_writer = csv.writer(new_file, delimiter=",")
                #for each paper in the list of papers it writes the rows into the appropriate place
                new_input = user_input()
                csv_writer.writerow(new_input)
        next_review()

    tk.Label(window, text="What is the course name? ").grid(row=0)
    tk.Label(window, text="What is the instructors name?").grid(row=1)
    tk.Label(window, text="Please enter your overall course rating from 1-5: ").grid(row=2)
    tk.Label(window, text="How difficult was the course? (1 being too easy and 5 being extremely difficult) ").grid(row=3)
    tk.Label(window, text="Would you retake the class?").grid(row=4)
    tk.Label(window, text="How was the workload? (1 being very light to 5 being too much)").grid(row=5)
    tk.Label(window, text="Were the exams/readings simple or intense? (1 being easiest to 5 being the hardest)").grid(row=6)
   

    e1 = tk.Entry(window)
    e2 = tk.Entry(window)
    e3= tk.Entry(window)
    e4= tk.Entry(window)
    e5= tk.Entry(window)
    e5= tk.Entry(window)
    e6= tk.Entry(window)
    e7= tk.Entry(window)

    e1.grid(row=0, column=1)
    e2.grid(row=1, column=1)
    e3.grid(row=2, column=1)
    e4.grid(row=3, column=1)
    e5.grid(row=4, column=1)
    e6.grid(row=5, column=1)
    e7.grid(row=6, column=1)
    tk.Button(window, text='Submit', command = add_to_csv).grid(row=7, column=1, sticky=tk.W, pady=4)
            
# to rename the title of the window 
root.title('Grade-A Course Finder')

# window dimensions
root.geometry('1600x1600')
root.configure(bg = "#000000")

label = tk.Label(root, text = 'Welcome to Grade-A Course Finder!',font = ('Eras bold itc', 50), fg = '#CF102D' ).pack()


a = tk.Button(root, text="Create Review", bg = '#FFCD23', height = 10, width = 20, command=create_review)
a['font'] = myFont
a.pack()

b = tk.Button(root, text="See Reviews", bg = '#FFCD23', height = 10, width = 20, command=see_review)
b['font'] = myFont
b.pack()


root.mainloop()
```
